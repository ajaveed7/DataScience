```python
import seaborn as sns
plt.rcParams['figure.figsize']=10,5 #this piece of code will set area of plots
```

```python
vis1 = sns.displot(stats['InternetUsers']) #distplot of univariate analysis
#uivariate analysis means only one attribute we have used in visualization
```

```python
#Distribution
vis2=sns.distplot(stats['BirthRate'])
```

```python
vis3=sns.distplot(stats['InternetUsers'],bins=20) #we have added 20 bins in this graph
```

```python
#Boxplot
vis4 = sns.boxplot(data = stats, x="IncomeGroup",y='BirthRate') #Bi-variate analysis
```

 in the above graph we can fiture out the following observations
    1. Income Group contains 4 types of groups
    2. their are outlisers in the data set except lower middle Income Group
    3. BirthRates are high in Low Income Group and very less in High Income Group
